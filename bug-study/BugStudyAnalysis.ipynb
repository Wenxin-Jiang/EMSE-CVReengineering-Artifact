```python
# Non-reengineering defects
sheet_n = pd.DataFrame(pd.read_csv('Non_Reengineering_Defects.csv'))
url = []
owner = []
name = []
issue_num = []
for i in sheet_n.iloc[:93, 0]:
  # j = 0
  # for i in sheet.iloc[:,:3].iloc[num]:
    # print(num, i) # name
    # if j == 0:
  url.append(i)
  # elif j == 1:
  owner.append(i.split('/')[3])
  # elif j == 2:
  name.append(i.split('/')[4])
  issue_num.append(i.split('/')[6])
  # j += 1
print(url[0:4], owner[0:4], name[0:4])
print(issue_num[0:4])

for i in range(len(owner)):
  res = []
  print(i)
  request = requests.get('https://api.github.com/repos/{}/{}/issues/{}'.format(owner[i], name[i], issue_num[i]), headers=headers)
  repo_request = requests.get('https://api.github.com/repos/{}/{}'.format(owner[i], name[i]), headers=headers)
  data = request.json()
  repo_data = repo_request.json()
  url = 'https://github.com/{}/{}/issues/{}'.format(owner[i], name[i], issue_num[i])
  try:
    created_time = data['created_at']
    repo_created = repo_data['created_at']
    sheet_n.iloc[i, 0] = url
    sheet_n.iloc[i, 1] = owner[i]
    sheet_n.iloc[i, 2] = name[i]
    sheet_n.iloc[i, 4] = issue_num[i]
    sheet_n.iloc[i, 7] = repo_created
    sheet_n.iloc[i, 8] = created_time
    sheet_n.iloc[i, 9] = (pd.to_datetime(created_time) - pd.to_datetime(repo_created)).days
    for j in range(len(data['labels'])):
      res.append(data['labels'][j]['name'])
    print(res)
    sheet_n.iloc[i, 10] = res
    # break
  except:
    pass

sheet_n.to_csv('NonReengineeringDefects.csv')
```
